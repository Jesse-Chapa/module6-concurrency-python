13.1 - Write the current date as a string to today.txt

(“Write the current date as a string to the text file today.txt.”) 

```python
from datetime import date

today = date.today().isoformat()  # ISO 8601 format (YYYY-MM-DD)
with open("today.txt", "w", encoding="utf-8") as f:
    f.write(today)

# verify
with open("today.txt", "r", encoding="utf-8") as f:
    print("today.txt contains:", f.read())

```

13.2 - Read today.txt into the string today_string

(“Read the text file today.txt into the string today_string.”)

```python
with open("today.txt", "r", encoding="utf-8") as f:
    today_string = f.read().strip()

today_string
```

13.3 - Parse the date from today_string

(“Parse the date from today_string.”)

```python
from datetime import date

parsed_date = date.fromisoformat(today_string)  # works because I wrote ISO format
parsed_date
```

15.1 — Use multiprocessing to create 3 processes

Each process should: wait random 0–1 seconds, print current time, exit.

```python
%%writefile mp_15_1.py
import multiprocessing as mp
import random
import time
from datetime import datetime

def worker(proc_num: int) -> None:
    wait_time = random.random()  # 0.0 to 1.0
    time.sleep(wait_time)
    print(f"Process {proc_num} waited {wait_time:.3f}s and prints time: {datetime.now().isoformat(timespec='seconds')}")

if __name__ == "__main__":
    procs = []
    for n in range(1, 4):  # three processes
        p = mp.Process(target=worker, args=(n,))
        p.start()
        procs.append(p)

    for p in procs:
        p.join()
```
